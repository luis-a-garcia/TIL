```python
def update_dice(pmf, data):
    """Update pmf based on new data."""
    hypos = pmf.qs
    likelihood = 1 / hypos
    impossible = hypos < data
    likelihood[impossible] = 0
    pmf *= likelihood
    pmf.normalize()
```

```python
pmf = prior.copy()
```

```python
update_dice(pmf, 1)
```

```python
update_dice(pmf, 3)
update_dice(pmf, 5)
update_dice(pmf, 7)
pmf
```
