```python
import pandas as pd
from fractions import Fraction

table = pd.DataFrame(
    index = ['6-sided', '8-sided', '12-sided'])

table['prior'] = [Fraction(1, 3), Fraction(1, 3), Fraction(1, 3)]
table['likelihood'] = [Fraction(1, 6), Fraction(1, 8), Fraction(1, 12)]
```

```python
def update(table):
    table['unnormalized'] = table['prior'] * table['likelihood']
    total = table['unnormalized'].sum()
    table['posterior'] = table['unnormalized'] / total
    return total    
```

```python
total = update(table)
```
