```python
def make_binomial(n, p):
    ''''make binomial Pmf'''
    ks = np.arange(n+1)
    ps = binom.pmf(ks, n, p)
    return Pmf(ps, ks)
```

```python
pmf_k = make_binomial(n=250, p=0.5)
```

```python
def prob_ge(pmf, threshold):
    """Probability of quantities great than threshold."""
    ge = (pmf.qs >= threshold)
    total = pmf[ge].sum()
    return total
```

```python
prob_ge(pmf_k, 140)
```
